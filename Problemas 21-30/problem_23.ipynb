```python
#Implementar y operar con matrices
import numpy as np

# Función para pedir una matriz al usuario
def pedir_matriz(filas, columnas):
    return np.array([[int(input(f"Ingrese el elemento en la posición ({i+1}, {j+1}): "))
                      for j in range(columnas)] for i in range(filas)])

# Función para realizar operaciones
def operar_matrices():
    while True:
        print("\nMenú de Operaciones con Matrices")
        print("1. Crear una matriz")
        print("2. Sumar matrices")
        print("3. Restar matrices")
        print("4. Multiplicar matrices")
        print("5. Transponer una matriz")
        print("6. Calcular el determinante de una matriz")
        print("7. Salir")

        opcion = input("Elige una opción: ")

        if opcion == '1':
            filas = int(input("Número de filas: "))
            columnas = int(input("Número de columnas: "))
            matriz = pedir_matriz(filas, columnas)
            print("Matriz creada:")
            print(matriz)

        elif opcion == '2':
            if 'matriz' in locals():
                filas = int(input("Número de filas de la segunda matriz: "))
                columnas = int(input("Número de columnas de la segunda matriz: "))
                matriz2 = pedir_matriz(filas, columnas)
                if matriz.shape == matriz2.shape:
                    print("Resultado de la suma:")
                    print(matriz + matriz2)
                else:
                    print("Las matrices no tienen las mismas dimensiones.")
            else:
                print("Primero crea una matriz para realizar operaciones.")

        elif opcion == '3':
            if 'matriz' in locals():
                filas = int(input("Número de filas de la segunda matriz: "))
                columnas = int(input("Número de columnas de la segunda matriz: "))
                matriz2 = pedir_matriz(filas, columnas)
                if matriz.shape == matriz2.shape:
                    print("Resultado de la resta:")
                    print(matriz - matriz2)
                else:
                    print("Las matrices no tienen las mismas dimensiones.")
            else:
                print("Primero crea una matriz para realizar operaciones.")

        elif opcion == '4':
            if 'matriz' in locals():
                filas = int(input("Número de filas de la segunda matriz: "))
                columnas = int(input("Número de columnas de la segunda matriz: "))
                matriz2 = pedir_matriz(filas, columnas)
                if matriz.shape[1] == matriz2.shape[0]:
                    print("Resultado de la multiplicación:")
                    print(np.dot(matriz, matriz2))
                else:
                    print("El número de columnas de la primera matriz debe ser igual al número de filas de la segunda.")
            else:
                print("Primero crea una matriz para realizar operaciones.")

        elif opcion == '5':
            if 'matriz' in locals():
                print("Matriz transpuesta:")
                print(matriz.T)
            else:
                print("Primero crea una matriz para realizar operaciones.")

        elif opcion == '6':
            if 'matriz' in locals():
                if matriz.shape[0] == matriz.shape[1]:  # La matriz debe ser cuadrada
                    print(f"El determinante de la matriz es: {np.linalg.det(matriz)}")
                else:
                    print("La matriz debe ser cuadrada para calcular el determinante.")
            else:
                print("Primero crea una matriz para realizar operaciones.")

        elif opcion == '7':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida.")

# Ejecutar el programa
operar_matrices()

import numpy as np

# Función para pedir una matriz al usuario
def pedir_matriz(filas, columnas):
    return np.array([[int(input(f"Ingrese el elemento en la posición ({i+1}, {j+1}): "))
                      for j in range(columnas)] for i in range(filas)])

# Función para realizar operaciones
def operar_matrices():
    while True:
        print("\nMenú de Operaciones con Matrices")
        print("1. Crear una matriz")
        print("2. Sumar matrices")
        print("3. Restar matrices")
        print("4. Multiplicar matrices")
        print("5. Salir")

        opcion = input("Elige una opción: ")

        if opcion == '1':
            filas = int(input("Número de filas: "))
            columnas = int(input("Número de columnas: "))
            matriz = pedir_matriz(filas, columnas)
            print("Matriz creada:")
            print(matriz)

        elif opcion == '2':
            if 'matriz' in locals():
                filas = int(input("Número de filas de la segunda matriz: "))
                columnas = int(input("Número de columnas de la segunda matriz: "))
                matriz2 = pedir_matriz(filas, columnas)
                if matriz.shape == matriz2.shape:
                    print("Resultado de la suma:")
                    print(matriz + matriz2)
                else:
                    print("Las matrices no tienen las mismas dimensiones.")
            else:
                print("Primero crea una matriz para realizar operaciones.")

        elif opcion == '3':
            if 'matriz' in locals():
                filas = int(input("Número de filas de la segunda matriz: "))
                columnas = int(input("Número de columnas de la segunda matriz: "))
                matriz2 = pedir_matriz(filas, columnas)
                if matriz.shape == matriz2.shape:
                    print("Resultado de la resta:")
                    print(matriz - matriz2)
                else:
                    print("Las matrices no tienen las mismas dimensiones.")
            else:
                print("Primero crea una matriz para realizar operaciones.")

        elif opcion == '4':
            if 'matriz' in locals():
                filas = int(input("Número de filas de la segunda matriz: "))
                columnas = int(input("Número de columnas de la segunda matriz: "))
                matriz2 = pedir_matriz(filas, columnas)
                if matriz.shape[1] == matriz2.shape[0]:
                    print("Resultado de la multiplicación:")
                    print(np.dot(matriz, matriz2))
                else:
                    print("El número de columnas de la primera matriz debe ser igual al número de filas de la segunda.")
            else:
                print("Primero crea una matriz para realizar operaciones.")
        elif opcion == '5':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida.")

# Ejecutar el programa
operar_matrices()
```
