```python
# Genera y analiza histogramas de datos

np.random.seed(0)
data = np.random.randn(1000)  # 1000 datos con distribución normal (media=0, std=1)
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
data = np.random.randn(1000)

# Crear el histograma
plt.hist(data, bins=20, edgecolor='skyblue', color='pink')

# Títulos
plt.title('Histograma de Datos Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Mostrar el histograma
plt.show()

#Análisis
print("Media de los datos: ", np.mean(data))
print("Desviación estándar de los datos: ", np.std(data))
print("Mediana de los datos: ", np.median(data))
print("Varianza de los datos: ", np.var(data))


```
