```python
#Generar y analizar datos estadisticos

import numpy as np
import matplotlib.pyplot as plt

# Genera datos aleatorios
def generar_datos(n, minimo, maximo):
    return np.random.randint(minimo, maximo, size=n)

# Calcula estadísticas básicas
def calcular_estadisticas(datos):
    media = np.mean(datos)
    mediana = np.median(datos)
    desviacion_estandar = np.std(datos)
    varianza = np.var(datos)

    return media, mediana, desviacion_estandar, varianza

# Graficar los datos
def graficar_datos(datos):
    plt.hist(datos, bins=10, color='pink', edgecolor='black')
    plt.title('Distribución de los Datos')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.show()

def main():
    n = 1000  # Número de datos
    minimo = 1  # Valor mínimo
    maximo = 100  # Valor máximo

    # Generamos los datos
    datos = generar_datos(n, minimo, maximo)

    media, mediana, desviacion_estandar, varianza = calcular_estadisticas(datos)

    print("Media: ", media)
    print("Mediana: ", mediana)
    print("Desviación Estándar: ", desviacion_estandar)
    print("Varianza: ", varianza)

    graficar_datos(datos)
main()
```
