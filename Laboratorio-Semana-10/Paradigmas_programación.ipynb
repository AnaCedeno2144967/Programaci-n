```python
#Paradigmas de programación

#Paradigma modular
def agregar_producto(inventario, nombre, cantidad, precio):
    inventario.append({'nombre': nombre, 'cantidad': cantidad, 'precio': precio})

def eliminar_producto(inventario, nombre):
    for producto in inventario:
        if producto['nombre'] == nombre:
            inventario.remove(producto)
            print("Producto ", nombre, "eliminado")
        else:
            print("Producto" ,nombre, "no encontrado")

def actualizar_producto(inventario, nombre, nueva_cantidad, nuevo_precio):
    for producto in inventario:
        if producto['nombre'] == nombre:
            producto['cantidad'] = nueva_cantidad
            producto['precio'] = nuevo_precio
            print("Producto ", nombre, "actualizado")
        else:
            print("Producto ", nombre, "no encontrado")

def ver_inventario(inventario):
    if len(inventario) == 0:
        return "El inventario está vacío."
    return "\n".join(["Producto: " + producto['nombre'] + " - Cantidad: " + str(producto['cantidad']) + " - Precio: " + str(producto['precio']) + " $" for producto in inventario])


# Clase que representa el inventario de la tienda (Paradigma Orientado a Objetos)
class InventarioTienda:
    def __init__(self):
        self.inventario = []

    def agregar(self, nombre, cantidad, precio):
        agregar_producto(self.inventario, nombre, cantidad, precio)

    def eliminar(self, nombre):
        return eliminar_producto(self.inventario, nombre)

    def actualizar(self, nombre, nueva_cantidad, nuevo_precio):
        return actualizar_producto(self.inventario, nombre, nueva_cantidad, nuevo_precio)

    def ver(self):
        return ver_inventario(self.inventario)

# Paradigma Estructurado
def gestion_inventario():
    tienda = InventarioTienda()

    while True:
        print("\nMenu de Opciones:")
        print("1. Agregar Producto")
        print("2. Eliminar Producto")
        print("3. Actualizar Producto")
        print("4. Ver Inventario")
        print("5. Salir")

        opcion = input("Elija una opción: ")
#Paradigma imperativo
        if opcion == '1':
            nombre = input("Nombre del producto: ")
            cantidad = int(input("Cantidad: "))
            precio = float(input("Precio: "))
            tienda.agregar(nombre, cantidad, precio)
            print("Producto ", nombre, "agregado al inventario")

        elif opcion == '2':
            nombre = input("Nombre del producto a eliminar: ")
            print(tienda.eliminar(nombre))

        elif opcion == '3':
            nombre = input("Nombre del producto a actualizar: ")
            nueva_cantidad = int(input("Nueva cantidad: "))
            nuevo_precio = float(input("Nuevo precio: "))
            print(tienda.actualizar(nombre, nueva_cantidad, nuevo_precio))

        elif opcion == '4':
            print("\nInventario de la Tienda:")
            print(tienda.ver())

        elif opcion == '5':
            print("Saliendo")
            break

        else:
            print("Opción no válida, intente de nuevo.")

gestion_inventario()
```
