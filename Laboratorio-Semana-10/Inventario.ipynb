```python
#Manejo de inventario con listas y diccionarios

#Utilicé una lista para guardar los productos
inventario = []

def agregar_producto():
  nombre = input("\nIngrese el nombre del producto: ")
  categoria= input("Ingrese la categoria del producto: ")
  precio = float(input("Ingrese el precio del producto: "))
  cantidad = int(input("Ingrese la cantidad del producto: "))
  #A cada producto corresponde un diccionario con información de si mismo
  producto = {
    "nombre": nombre,
    "categoria": categoria,
    "precio": precio,
    "cantidad": cantidad
  }
  # append() se utiliza para agregar elementos
  inventario.append(producto)
  print("Producto agregado")
  opcion = input("Elija 1 volver al menú o presione cualquier tecla para salir: ")
  if opcion == "1":
    menu()
  else:
    print("Saliendo")

def eliminar_producto():
  nombre = input("\nIngrese el nombre del producto a eliminar: ")
  encontrado = False
# Recorremos producto por producto en la lista para buscar elelemento
  for producto in inventario:
    if producto["nombre"].lower() == nombre.lower():
# remove() para eliminar elementos
      inventario.remove(producto)
      print("Producto eliminado")
      encontrado = True
      break
  if not encontrado:
    print("Producto no existente")
  opcion = input("Elija 1 para volver al menú o presione cualquier tecla para salir: ")
  if opcion == "1":
    menu()
  else:
    print("Saliendo")

def buscar_producto():
  nombre = input("\nIngrese el nombre del producto a buscar: ")
  encontrado = False
  for producto in inventario:
    if producto["nombre"].lower() == nombre.lower():
      print("producto:", producto["nombre"],
      "categoria: ", producto["categoria"],
      "precio: ", producto["precio"],
      "cantidad: ", producto["cantidad"])
      encontrado = True
      break
  if not encontrado:
     print("Producto no existente")
  opcion = input("Elija 1 para volver al menú o presione cualquier tecla para salir: ")
  if opcion == "1":
    menu()
  else:
    print("Saliendo")


def mostrar_productos_ordenados():
#sorted() se usa para ordenar
  if inventario:
    productos_ordenados = sorted(inventario, key=lambda x: x["precio"])
    for producto in productos_ordenados:
       print("\nproducto:", producto["nombre"],
       " categoria:", producto["categoria"],
       " precio:", producto["precio"],
       " cantidad:", producto["cantidad"])
  else:
    print("El inventario está vacío")
  opcion = input("Elija 1 para volver al menú o presione cualquier tecla para salir: ")
  if opcion == "1":
      menu()
  else:
      print("Saliendo")

def menu():
  print("\nMenú de inventario")
  print("1.- Agregar producto")
  print("2.- Eliminar producto")
  print("3.- Buscar productos")
  print("4.- Mostrar productos ordenados por precio")

  opcion = int(input("Ingrese una opción: "))

  if opcion == 1:
    agregar_producto()
  elif opcion == 2:
    eliminar_producto()
  elif opcion == 3:
    buscar_producto()
  elif opcion == 4:
    mostrar_productos_ordenados()


menu()

```
